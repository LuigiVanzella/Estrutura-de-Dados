Exercicio 2

```python
class Node:
  def __init__(self, valor):
    self.valor = valor
    self.filhos = []

raiz = Node('Avô')
print("Raiz:", raiz.valor)

no_pai = Node("Pai")
raiz.filhos = [no_pai]
print("Pai:", no_pai.valor)

no_filho1 = Node("Filho 1")
no_filho2 = Node("Filho 2")
no_pai.filhos = [no_filho1, no_filho2]
print("familia: Raiz:", raiz.valor, ", Pai:", no_pai.valor, ", Filhos:", [filhos.valor for filhos in no_pai.filhos] )

```

```python
def imprimirSubArvore(no):
  print(no.valor, end=" ")
  for filho in no.filhos:
    imprimirSubArvore(filho)

print("SubArvore de Pai:", end= " ")
imprimirSubArvore(no_pai)
```

```python

raiz = Node("A")

no_b = Node("B")
no_c = Node("C")
raiz.filhos = [no_b, no_c]

no_d = Node("D")
no_c.filhos = [no_d]

def altura(no):
  if not no.filhos:
    return 1
  return 1 + max(altura(f) for f in no.filhos)

print("Altura da árvore:", altura(raiz))
```

Exercicio 5

```python
def profundidade(raiz, alvo, nivel = 0):
  if raiz.valor == alvo:
    return nivel
  for filho in raiz.filhos:
    prof = profundidade(filho, alvo, nivel + 1)
    if prof is not None:
      return prof
  return None

print("Profundidade do nó D:", profundidade(raiz, "D"))
```

Exercicio 7

```python
raiz = Node("A")

no_b = Node("B")
no_c = Node("C")
raiz.filhos = [no_b, no_c]

no_d = Node("D")
no_e = Node("E")
no_f = Node("F")

no_b.filhos = [no_d, no_e]
no_c.filhos = [no_f]

def caminhos_possiveis(raiz):
  if raiz is None:
    return []
  if not raiz.filhos:
    return [[raiz.valor]]
  caminhos = []
  for filho in raiz.filhos:
    for caminho in caminhos_possiveis(filho):
      caminhos.append([raiz.valor] + caminho)

  return caminhos

def listar_caminhos(raiz):
  for cam in caminhos_possiveis(raiz):
    print(" -> ".join(map(str, cam)))

listar_caminhos(raiz)
```

Exercicio 8

```python
def contar_folhas(raiz):
  if not raiz.filhos:
    return 1
  count_leaves = 1
  for filho in raiz.filhos:
    if filho.filhos:
      contar_folhas(filho)

    count_leaves = count_leaves + 1

  return count_leaves

print("Total de folhas", contar_folhas(raiz))
```

Exercicio 9


```python
def sao_irmaos(raiz, x, y):
  if not raiz.filhos:
    return False

  nomes_dos_filhos = [filho.valor for filho in raiz.filhos]

  if x in nomes_dos_filhos and y in nomes_dos_filhos:
    return True

  for filho in raiz.filhos:
    if sao_irmaos(filho, x, y):
      return True
  return False

print("São irmãos:", sao_irmaos(raiz, "D", "E"))
```
