# Parte 1 — Bubble, Selection e Insertion Sort
Este notebook contém exercícios diretos para treinar os três algoritmos.

## Código base (use estas funções como referência para testar)

```python
# Você pode editar, completar e usar para testar seus exercícios

def bubble_sort(v):
    v = v[:]
    comp, troc = 0, 0
    for i in range(len(v)-1):
        for j in range(len(v)-i-1):
            comp += 1
            if v[j] > v[j+1]:
                v[j], v[j+1] = v[j+1], v[j]
                troc += 1
    return v, (comp, troc)

def selection_sort(v):
    v = v[:]
    comp, troc = 0, 0
    for i in range(len(v)-1):
        m = i
        for j in range(i+1, len(v)):
            comp += 1
            if v[j] < v[m]:
                m = j
        if m != i:
            v[i], v[m] = v[m], v[i]
            troc += 1
    return v, (comp, troc)

def insertion_sort(v):
    v = v[:]
    comp, desl = 0, 0
    for i in range(1, len(v)):
        chave = v[i]
        j = i - 1
        while j >= 0:
            comp += 1
            if v[j] > chave:
                v[j+1] = v[j]
                desl += 1
                j -= 1
            else:
                break
        v[j+1] = chave
    return v, (comp, desl)
```
